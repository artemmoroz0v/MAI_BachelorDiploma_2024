```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator
from matplotlib.patches import Patch
```

```python
df = pd.read_excel("data_dima.xlsx")
```

# Графики по оценкам для четырех типов семей

**Зависимость оценок по русскому языку и математики от типа семьи**

```python
average_rus_by_family = df.groupby('Family')['Rus'].mean().sort_values().reset_index()
average_math_by_family = df.groupby('Family')['Math'].mean().sort_values().reset_index()
sns.set(style="whitegrid")
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))
sns.barplot(x='Family', y='Rus', data=average_rus_by_family, order=average_rus_by_family['Family'], palette='viridis', ax=axes[0])
axes[0].set(xlabel=None, ylabel=None)
axes[0].set_title('Средняя оценка по русскому языку в зависимости от типа семьи', fontsize=14)
axes[0].set_ylim(3.6, 3.85)
for index, value in enumerate(average_rus_by_family['Rus']):
    axes[0].text(index, value + 0.0015, f'{value:.3f}', ha='center', va='bottom', color='black', fontsize=10, fontweight='bold')
axes[0].grid(axis='y', linestyle='--', alpha=0.9)
axes[0].set_facecolor('#f5f5f5')
axes[0].spines['bottom'].set_color('black')
axes[0].spines['top'].set_color('black')
axes[0].spines['right'].set_color('black')
axes[0].spines['left'].set_color('black')
axes[0].tick_params(axis='x', colors='black')
axes[0].tick_params(axis='y', colors='black')
sns.barplot(x='Family', y='Math', data=average_math_by_family, order=average_math_by_family['Family'], palette='Set2', ax=axes[1])
axes[1].set(xlabel=None, ylabel=None)
axes[1].set_title('Средняя оценка по математике в зависимости от типа семьи', fontsize=15)
axes[1].set_ylim(3.6, 3.85)
for index, value in enumerate(average_math_by_family['Math']):
    axes[1].text(index, value + 0.0015, f'{value:.3f}', ha='center', va='bottom', color='black', fontsize=10, fontweight='bold')
axes[1].grid(axis='y', linestyle='--', alpha=0.9)
axes[1].set_facecolor('#f5f5f5')
axes[1].spines['bottom'].set_color('black')
axes[1].spines['top'].set_color('black')
axes[1].spines['right'].set_color('black')
axes[1].spines['left'].set_color('black')
axes[1].tick_params(axis='x', colors='black')
axes[1].tick_params(axis='y', colors='black')
plt.show()
```
